```python
#(a) Loading the dataset

from keras.utils import to_categorical
from keras.datasets import mnist

(train_X, train_Y), (test_X, test_Y) = mnist.load_data ()
```

```python
#b Preprocessing

print(train_X.shape)
print(train_Y.shape)
print("")
print(test_X.shape)
print(test_Y.shape)
```

```python
# adding a dimension to the training data 
train_X_reshaped = train_X.reshape(train_X.shape[0],28,28,1)
```

```python
#transformed shape of training data after adding another dimension

train_X_reshaped.shape
```

```python
# adding a dimension to the training data 

test_X_reshaped = test_X.reshape(test_X.shape[0],28,28,1)
```

```python
#transformed shape of test data after adding another dimension

test_X_reshaped.shape
```

```python
# scaling pixel values to ensure they all lie between 0 and 1
train_X_reshaped = train_X_reshaped/255
test_X_reshaped = test_X_reshaped/255
```

```python
train_Y[1:10,]
```

```python
from keras.utils import to_categorical

train_labels = to_categorical(train_Y, num_classes = 10, dtype = 'int')
train_labels[1:10,]
```

```python
#printing our current values of the test set labels 
test_Y[1:10,]
```

```python
# converting test labels to a one-hot representation

test_labels = to_categorical(test_Y, num_classes = 10, dtype = 'int')
test_labels[1:10,]
```

```python
# c) Implementation

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD



def create_cnn():
    # define using Sequential
    model = Sequential()
    
    model.add(
    Conv2D(32, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    #Maxpooling layer
    model.add(MaxPooling2D((2,2)))
    
    #Flatten output
    model.add(Flatten())
    
    #Desne layer of 100 neurons
    model.add(
        Dense(100,
        activation = 'relu', 
        kernel_initializer = 'he_uniform')
    )
    model.add(Dense(10, activation = 'softmax'))
    
    #initialize optimizer
    opt = SGD(learning_rate = .01, momentum = 0.9)
    
    #compile model
    model.compile(
        optimizer = opt,
        loss = 'categorical_crossentropy',
        metrics = ['accuracy']
    )
    
    return model
```

```python
# d) Training and Evaluating CNN

model = create_cnn()
model.fit(train_X_reshaped, train_labels, batch_size = 32, epochs =10, validation_split =0.1)
score = model.evaluate(test_X_reshaped, test_labels, verbose =0)
```

```python
#e) Experimentation

# 50 epochs

epoch_history = model.fit(train_X_reshaped, train_labels, batch_size = 32, epochs=50, validation_split = 0.1)

# print validation and training accuracy over epochs
print (epoch_history.history[ 'accuracy' ] )
print (epoch_history.history[ 'val_accuracy' ] )
```

```python
# e) iii. Add another convolution layer and maxpooling layer 

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD



def create_cnn_added_layers():
    # define using Sequential
    model = Sequential()
    
    model.add(
    Conv2D(32, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    #Maxpooling layer
    model.add(MaxPooling2D((2,2)))
    
    #Added convolutional layer
    model.add(
    Conv2D(64, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    model.add(MaxPooling2D((2,2)))

    
    #Flatten output
    model.add(Flatten())
    
    #Desne layer of 100 neurons
    model.add(
        Dense(100,
        activation = 'relu', 
        kernel_initializer = 'he_uniform')
    )
    model.add(Dense(10, activation = 'softmax'))
    
    #initialize optimizer
    opt = SGD(learning_rate = .01, momentum = 0.9)
    
    #compile model
    model.compile(
        optimizer = opt,
        loss = 'categorical_crossentropy',
        metrics = ['accuracy']
    )
    
    return model
```

```python
model_added_layers = create_cnn_added_layers()
```

```python
#2. Random Forests for Image Approximation 

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
%matplotlib inline

sample_size = 5000

img = Image.open("aml_hw4/monaalisa.jpg","r")

img = np.asarray(img)
plt.imshow(img)
plt.xlabel("Original")
plt.show()
```

```python
# process data

data, label = [], []
row = img.shape[0]
col = img.shape[1]
for i in range(sample_size):
    point = []
    point.append(np.random.randint(row))
    point.append(np.random.randint(col))
    data.append(point)
    label.append(img[point[0],point[1]])
data = np.array(data)
label = np.array(label)
```

```python
# create a white image
sampledImg = np.zeros([row,col,3], dtype=np.uint8)
sampledImg.fill(255)
for i in range(sample_size):
    sampledImg[data[i,0]][data[i,1]] = label[i]
plt.imshow(sampledImg)
plt.xlabel("Sampled Points")
plt.show()
```

```python
# 2d. Rescale the pixel intensities.
scaledImg = np.array(img, dtype=float)
for i in range(row):
    for j in range(col):
        scaledImg[i,j] = img[i,j] / 255
plt.imshow(scaledImg)
plt.xlabel("Scaled Image")
plt.show()

plt.xlabel("Original")
plt.imshow(img)
plt.show()

scaledLabel = np.array(label, dtype=float)
scaledLabel /= 255    
```
