```python
#(a) Loading the dataset

from keras.utils import to_categorical
from keras.datasets import mnist

(train_X, train_Y), (test_X, test_Y) = mnist.load_data ()
```

```python
# adding a dimension to the training data 
train_X_reshaped = train_X.reshape(train_X.shape[0],28,28,1)
```

```python
# adding a dimension to the training data 

test_X_reshaped = test_X.reshape(test_X.shape[0],28,28,1)
```

```python
#transformed shape of test data after adding another dimension

test_X_reshaped.shape
```

```python
# scaling pixel values to ensure they all lie between 0 and 1
train_X_reshaped = train_X_reshaped/255
test_X_reshaped = test_X_reshaped/255
```

```python
train_Y[1:10,]
```

```python
from keras.utils import to_categorical

train_labels = to_categorical(train_Y, num_classes = 10, dtype = 'int')
train_labels[1:10,]
```

```python
#printing our current values of the test set labels 
test_Y[1:10,]
```

```python
# converting test labels to a one-hot representation

test_labels = to_categorical(test_Y, num_classes = 10, dtype = 'int')
test_labels[1:10,]
```

```python
# c) Implementation

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD



def create_cnn():
    # define using Sequential
    model = Sequential()
    
    model.add(
    Conv2D(32, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    #Maxpooling layer
    model.add(MaxPooling2D((2,2)))
    
    #Flatten output
    model.add(Flatten())
    
    #Desne layer of 100 neurons
    model.add(
        Dense(100,
        activation = 'relu', 
        kernel_initializer = 'he_uniform')
    )
    model.add(Dense(10, activation = 'softmax'))
    
    #initialize optimizer
    opt = SGD(learning_rate = .01, momentum = 0.9)
    
    #compile model
    model.compile(
        optimizer = opt,
        loss = 'categorical_crossentropy',
        metrics = ['accuracy']
    )
    
    return model
```

```python
# d) Training and Evaluating CNN

model = create_cnn()
model.fit(train_X_reshaped, train_labels, batch_size = 32, epochs =10, validation_split =0.1)
score = model.evaluate(test_X_reshaped, test_labels, verbose =0)
```

```python
#e) Experimentation

# 50 epochs

epoch_history = model.fit(train_X_reshaped, train_labels, batch_size = 32, epochs=50, validation_split = 0.1)

# print validation and training accuracy over epochs
print (epoch_history.history[ 'accuracy' ] )
print (epoch_history.history[ 'val_accuracy' ] )
```

```python
# e) iii. Add another convolution layer and maxpooling layer 

from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.optimizers import SGD



def create_cnn_added_layers():
    # define using Sequential
    model = Sequential()
    
    model.add(
    Conv2D(32, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    #Maxpooling layer
    model.add(MaxPooling2D((2,2)))
    
    #Added convolutional layer
    model.add(
    Conv2D(64, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    model.add(MaxPooling2D((2,2)))

    
    #Flatten output
    model.add(Flatten())
    
    #Desne layer of 100 neurons
    model.add(
        Dense(100,
        activation = 'relu', 
        kernel_initializer = 'he_uniform')
    )
    model.add(Dense(10, activation = 'softmax'))
    
    #initialize optimizer
    opt = SGD(learning_rate = .01, momentum = 0.9)
    
    #compile model
    model.compile(
        optimizer = opt,
        loss = 'categorical_crossentropy',
        metrics = ['accuracy']
    )
    
    return model
```

```python
model_added_layers = create_cnn_added_layers()
```

```python
model_added_layers.fit(train_X_reshaped, train_labels, batch_size = 32, epochs =10, validation_split =0.1)
score = model_added_layers.evaluate(test_X_reshaped, test_labels, verbose =0)
```

```python
#e) iv. 


def create_cnn_iv():
    # define using Sequential
    model = Sequential()

    model.add(Dropout(0.5))


    model.add(
    Conv2D(32, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    #Maxpooling layer
    model.add(MaxPooling2D((2,2)))
    
    #Added convolutional layer
    model.add(
    Conv2D(64, (3,3), 
           activation = 'relu',
           kernel_initializer = 'he_uniform',
           input_shape = (28,28,1))
    )
    
    model.add(MaxPooling2D((2,2)))

    
    #Flatten output
    model.add(Flatten())
    
    #Desne layer of 100 neurons
    model.add(
        Dense(100,
        activation = 'relu', 
        kernel_initializer = 'he_uniform')
    )
    model.add(Dense(10, activation = 'softmax'))
    
    #initialize optimizer
    opt = SGD(learning_rate = .1, momentum = 0.9)
    
    #compile model
    model.compile(
        optimizer = opt,
        loss = 'categorical_crossentropy',
        metrics = ['accuracy']
    )
    
    return model
```

#### e) Experimentation. 
#### i. Repeat the experiment for a random forest containing a single decision tree, but with depths 1, 2, 3, 5, 10, and 15. How does depth impact the result? Describe in detail why.


As the depth increases from 1 - 15 in varying increments, while keeping constant the number of decision trees at one, the results become noticebly better. As the depth increases, the model is able to gather more information and intuition about how to appropiately places colored pixels. This increased accuracy is enabled because the model is able to gather more information about the data by increasing the number of splits. While the number of layers is low, as in this case, the model is underfitting based on the training data pixel coordinates. Conversely, if depth number is raised too high, the model will be overfited based on the training data pixel coordinates. There is no maximum number to assign the depth to, rather it's dependent on the image in which the random forest algorithm is being applied to.
